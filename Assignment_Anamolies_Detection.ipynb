Q1. What is Dimensionality Reduction? Why is it important in machine learning?

Dimensionality reduction is the process of reducing the number of features (or variables) in a dataset while keeping the most important information intact. It is important in machine learning because it helps prevent the curse of dimensionality, which can lead to slower training, overfitting, and poor model performance by making data sparser and patterns harder to detect in high-dimensional spaces. This simplification improves model efficiency, reduces computational complexity, and makes it easier to visualize and understand the data. 

* Why dimensionality reduction is important in machine learning

Reduces computational complexity: Less features means less data to process, which leads to faster training times and reduced computational resource needs.

Prevents overfitting: By removing redundant or irrelevant features, it helps the model focus on the most important patterns, making it less likely to overfit the training data and improving its ability to generalize to new, unseen data.

Simplifies the model: A model with fewer features is often easier to understand and interpret.

Improves data visualization: It's difficult to visualize data in more than three dimensions, so reducing the number of dimensions allows for more effective visualization and analysis of the data's structure.

Reduces storage space: A lower-dimensional dataset requires less storage space.

Removes noise: Techniques like Principal Component Analysis (PCA) can remove noise and irrelevant features by finding a new, smaller set of variables that capture the most important visual patterns or statistical information. 

Q2. Name and briefly describe three common dimensionality reduction techniques

Three common dimensionality reduction techniques are

1. Principal Component Analysis (PCA)

Description: A linear, unsupervised technique that transforms the original features into a new, smaller set of uncorrelated variables called principal components.

How it works: It finds new axes (principal components) that capture the maximum variance in the data. The first component captures the most variance, the second captures the next most, and so on. By keeping the first few components, you can reduce the dimensionality while retaining most of the original information. 

2. Linear Discriminant Analysis (LDA)

Description: A supervised, linear technique that aims to reduce the dimensionality of data while maximizing the separability between classes.

How it works: Unlike PCA which is unsupervised, LDA uses class labels to find a new feature space that better separates the different categories in the dataset. 

3. t-Distributed Stochastic Neighbor Embedding (t-SNE)

Description: A non-linear technique primarily used for visualizing high-dimensional data in 2D or 3D.

How it works: It works by converting the high-dimensional Euclidean distances between data points into conditional probabilities, and then trying to match these probabilities to a low-dimensional embedding. It excels at revealing local structure and clusters in the data.

Q3. What is clustering in unsupervised learning? Mention three popular clustering algorithms.

* Clustering in Unsupervised Machine Learning

Clustering is an unsupervised machine learning technique used to group similar data points into clusters without predefined labels. It identifies hidden patterns or structures in data by measuring the similarity or distance between points. Data points within the same cluster share common characteristics, while those in different clusters are distinct. Clustering is widely used in customer segmentation, market analysis, image processing, and anomaly detection. Common algorithms include K-Means, Hierarchical Clustering, and DBSCAN. Overall, clustering helps simplify complex datasets, uncover natural groupings, and support better decision-making by revealing meaningful relationships in unlabeled data.

K-Means: An algorithm that partitions data into k clusters by minimizing the distance between each data point and its assigned cluster's centroid.

Hierarchical Clustering: Builds a tree of clusters. It can start with each data point as its own cluster and then iteratively merge the closest clusters, or start with a single cluster and split it recursively.

DBSCAN: A density-based algorithm that groups together points that are closely packed, marking as outliers the points in low-density regions.

Q4. Explain the concept of anomaly detection and its significance.


Anomaly Detection (also known as Outlier Detection) is a technique used in machine learning and data analysis to identify data points, events, or observations that deviate significantly from the normal behavior of a dataset.

* Concept of anomaly detection

Identification of outliers: It involves finding data points that are statistically different from the rest of the dataset, often referred to as outliers or novelties.

Deviation from normal patterns: The core idea is to define what "normal" looks like and then flag anything that doesn't conform to that expected behavior.

Application of techniques: Various methods, including statistical and machine learning algorithms, are used to analyze data and identify these anomalies. 

* Significance and applications

Cybersecurity: Detects suspicious activity, network breaches, and cyberattacks by flagging unusual network traffic or login patterns.

Fraud prevention: Identifies fraudulent transactions in financial systems by spotting unusual spending patterns or account activities.

Manufacturing and quality control: Monitors equipment for signs of failure or defects in products by analyzing sensor data like temperature and vibration, allowing for proactive maintenance and quality assurance.

Healthcare: Can be used to detect unusual symptoms or patterns in patient data that may indicate a health issue.

Business insights: Uncovers unexpected opportunities, such as a product or salesperson outperforming expectations, which can be investigated further to understand what's working well.

IT system monitoring: Helps ensure operational efficiency by detecting unusual patterns in server logs that could signal a fault or failure. 

Q5. List and briefly describe three types of anomaly detection techniques.

1. Statistical Methods
These methods assume that normal data follow a specific probability distribution (such as a normal distribution).
Data points that fall far from the expected range or have low probability are marked as anomalies.


Example: Z-score, Gaussian model.

Use case: Detecting unusual temperature or sales data.

2. Machine Learning–Based Methods
These techniques use algorithms to learn normal behavior from data and detect deviations.
They can be supervised, unsupervised, or semi-supervised.

Example: Isolation Forest, One-Class SVM, Autoencoders.

Use case: Fraud detection, network intrusion detection.

3. Clustering-Based Methods
These methods group similar data points together into clusters.
Points that do not belong to any cluster or are far from cluster centers are considered anomalies.

Example: K-Means, DBSCAN.

Use case: Customer segmentation, identifying unusual user behavior.

Q6. What is time series analysis? Mention two key components of time series data

Time series analysis is the process of examining a sequence of data points ordered over time to identify trends, patterns, and seasonality, with the goal of forecasting future outcomes. Two key components of time series data are the trend, which is the long-term upward or downward movement, and seasonal variation, which is a predictable, repeating pattern within a fixed period, such as annual or monthly cycles. 

Key components of time series data

Trend: The long-term, overall direction of the data over a long period. For example, a company's sales might show an upward trend over a decade, or the price of a stock might have a downward trend over a specific period.

Seasonal Variation: Patterns that repeat at fixed intervals of time, such as daily, weekly, monthly, or yearly. An example is retail sales increasing during holiday seasons.


Q7. Describe the difference between seasonality and cyclic behavior in time series.


Difference Between Seasonality and Cyclic Behavior in Time Series

Both seasonality and cyclic behavior represent patterns that repeat over time in a time series, but they differ mainly in duration and predictability.

* Seasonality

Definition: Seasonality refers to regular and predictable patterns that occur at fixed time intervals — such as daily, monthly, or yearly.

Cause: Usually due to calendar-related factors (like weather, holidays, or business cycles).

Duration: Occurs over a short, fixed period.

Example:

Increased ice cream sales every summer.

Higher electricity usage during winter months.

* Cyclic Behavior

Definition: Cyclic behavior refers to fluctuations that occur over longer, irregular time periods and are often related to economic or business cycles.

Cause: Driven by non-seasonal factors like economic growth, recessions, or industry trends.

Duration: Occurs over a variable and usually longer period (often several years).

Example:

Economic booms and recessions.

Housing market ups and downs.

Q8. Write Python code to perform K-means clustering on a sample dataset.

```python
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data 


kmeans = KMeans(n_clusters=3, random_state=1)
kmeans.fit(X)

labels = kmeans.labels_
centers = kmeans.cluster_centers_

plt.scatter(X[:, 2], X[:, 3], c = labels, cmap='viridis', s = 50)
plt.scatter(centers[:, 2], centers[:, 3], c = 'red', s = 200, marker='X')
plt.xlabel('petal Length')
plt.ylabel('Petal Width')
plt.title('K-Means Clustering on Iris Dataset')
plt.show()

```

Q9. What is inheritance in OOP? Provide a simple example in Python.

Inheritance in Object-Oriented Programming (OOP) is a mechanism that allows a new class (child class or derived class) to inherit attributes and methods from an existing class (parent class or base class). This promotes code reusability and establishes a hierarchical relationship between classes, often referred to as an "is-a" relationship (e.g., a "Dog is an Animal"). 

```python
# Parent class
class Animal:
    def speak(self):
        print("This animal makes a sound.")

# Child class inheriting from Animal
class Dog(Animal):
    def speak(self):  # Overriding the parent method
        print("The dog barks.")

# Create objects
animal = Animal()
dog = Dog()

# Call methods
animal.speak()   
dog.speak() 
```

Q10. How can time series analysis be used for anomaly detection?

Time series analysis detects anomalies by modeling normal behavior and identifying data points that deviate significantly from it, using methods like seasonal decomposition (identifying outliers in the residual component) or machine learning models (such as autoencoders or isolation forests) that learn patterns and flag data with high reconstruction error or short path lengths. The core idea is to use the analysis of normal patterns—trends, seasonality, and cyclical behavior—to flag events that don't fit the established baseline. 

* Traditional and statistical methods

Seasonal decomposition: Breaks a time series into its constituent parts: trend, seasonality, and residuals (or remainder). Anomalies are detected as points with unusually large values in the residual component.

Statistical thresholds: Uses methods like z-score analysis or moving averages to establish a range for expected values. Points falling outside this range are flagged as anomalies. 

* Machine learning methods

Autoencoders: These are neural networks trained to reconstruct typical time series data. When an anomaly is fed into the trained autoencoder, the reconstruction error (the difference between the original and reconstructed data) will be large, indicating it's an anomaly.

Isolation Forest: This method works by randomly partitioning the data. Anomalies are isolated with shorter paths in the resulting tree structure, making them easier and faster to find than normal data points.

Deep learning models: More advanced models like LSTMs, Transformers, and Graph Neural Networks can be used to capture complex temporal dynamics and interdependencies in high-dimensional or multivariate time series data.

Unsupervised and semi-supervised learning: Many methods learn from unlabeled data. They are trained on "normal" data and flag anything that doesn't conform to these learned patterns.
